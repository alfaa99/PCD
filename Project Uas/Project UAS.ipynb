```python
#Ubah RGB ke Grayscale
import numpy as np
import imageio
import matplotlib.pyplot as plt
```

```python
img = imageio.imread("melati.jpg")
```

```python
img_height = img.shape[0]
img_width = img.shape[1]
img_channel = img.shape[2]
img_type = img.dtype
```

```python
img_gray = np.zeros(img.shape, dtype=np.uint8)
```

```python
for y in range(0, img_height):
    for x in range(0, img_width):
        red = img[y][x][0]
        green = img[y][x][1]
        blue = img[y][x][2]
        gray = (int(red) + int(green) + int(blue)) / 3
        img_gray[y][x] = (gray, gray, gray)
```

```python
plt.imshow(img_gray)
plt.imsave("grey.png", img_gray)
plt.title("Grayscale")
plt.show()
```

```python
#Tahap Cropping
from PIL import Image
from PIL import ImageFont
from PIL import ImageDraw
import sys
 
in_file = 'grey.png'
 
img = Image.open(in_file)
width, height = img.size
 
# crop
# 10 pixels from the left
# 20 pixels from the top
# 30 pixels from the right
# 40 pixels from the bottom
 
 
cropped = img.crop((400, 30, width-30, height-40))
cropped.save('greycrop.png')
```

```python
#membuat histogram
hg = np.zeros((256))
for x in range(0, 256):
    hg[x] = 0
    
for y in range(0, img_height):
    for x in range(0, img_width):
        gray = img_gray[y][x][0]
        hg[gray] += 1
        
bins = np.linspace(0, 256, 100)
plt.hist(hg, bins, color="black", alpha=0.5)
plt.title("Histogram")
plt.show()
```
