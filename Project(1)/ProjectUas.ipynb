# Ubah RGB ke Grayscale

# Bunga Melati

```python
import numpy as np
import imageio
import matplotlib.pyplot as plt
```

```python
img = imageio.imread("melati.jpg")
```

```python
#mendapatkan resolusi dan type dari gambar
img_height = img.shape[0]
img_width = img.shape[1]
img_channel = img.shape[2]
img_type = img.dtype
```

```python
img_gray = np.zeros(img.shape, dtype=np.uint8)
```

```python
for y in range(0, img_height):
    for x in range(0, img_width):
        red = img[y][x][0]
        green = img[y][x][1]
        blue = img[y][x][2]
        gray = (int(red) + int(green) + int(blue)) / 3
        img_gray[y][x] = (gray, gray, gray)
```

```python
plt.imshow(img_gray)
plt.imsave("melatigrey.jpg", img_gray)
plt.title("Grayscale")
plt.show()
```
